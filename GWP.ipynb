```python
print("QUESTION 3A: FIRST APPORACH: CHECKING ALL COMBINATIONS AND FINDINNG BEST") 

import pandas as pd
import statsmodels.api as sm
import itertools


df = pd.read_csv("FE-GWP1_model_selection_2.csv")
df.columns = df.columns.str.strip() 


y = df['Y']
predictors = ['Z1', 'Z2', 'Z3', 'Z4', 'Z5']

def fit_model(cols):
    
    if len(cols) == 0:
        X = pd.DataFrame(1, index=df.index, columns=['intercept'])
    else:
        X = df[list(cols)]
        X = sm.add_constant(X) 
    
    model = sm.OLS(y, X).fit()
    
    return {
        "Model": " + ".join(cols) if cols else "Intercept only",
        "Adj_R2": model.rsquared_adj,
        "AIC": model.aic,
        "BIC": model.bic
    }


results = []

for r in range(len(predictors) + 1):
    for combo in itertools.combinations(predictors, r):
        results.append(fit_model(combo))


results_df = pd.DataFrame(results)


print("Best Models by Adjusted R-squared (Highest is best) , AIC and BIC (Lowest is best")
print(results_df.sort_values(by="Adj_R2", ascending=False).head(3))
```
